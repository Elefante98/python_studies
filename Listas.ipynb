```python
#Listas são conjuntos ordenados e mutáveis de dados
lista_nomes = ["Moisés", "Nicolly", "Maraia", "Laudmilla"]
#São criadas usando um nome que recebe seus valores entre colchetes


#Acessamos os valores de uma lista pelo seu índice
print(f"Esse é o primeiro item da lista: {lista_nomes[0]}")

print(f"Esse é o segundo item da lista: {lista_nomes[1]}")

#Para acessarmos o último item de uma lista, mesmo que não saibamos quantos itens a lista tenha, usamos o índice -1
print(f"Esse é o último item da lista: {lista_nomes[-1]}")

print(f"Esse é o penúltimo item da lista: {lista_nomes[-2]}")
```

```python
#Operações básicas com listas
lista1 = ["Moisés", "Marcão", "Bea", "Dynamike", "Super Xandão"]

#Adicionar elementos: append() e insert()

lista1.append("Buster") #Adiciona a string Buster ao final da lista

lista1.insert(1, "Jacky") #Adiciona a string Jacky no índice indicado
print(lista1)
print()





#Remover elementos: remove() e pop()

lista1.remove("Bea") #Remove o elemento passado como argumento

lista1.pop() #Remove o índice especificado, ou, se não for especificado, o último elemento da lista






#Concatenar listas: "+" e extend()

lista2 = [1, 2, 3, 4, 5]
lista3 = [6, 7, 8, 9, 10]
#O "+" junta as listas em uma única
listona = lista2 + lista3

print(f"Estas listas foram concatenadas com '+' {listona}")
print()

lista2.extend(lista3)

print(f"Já essas, foram concatenadas com .extend() {lista2}")
print()






#Repetir listas: *

numbers = [1, 2, 3, 4, 5]

print(numbers * 4) #O "*" repete a lista a quantidade de vezes estipulada após ele
print()






#Verificar se um item está na lista: in
nomes = ["Moisés", "Nicolly", "Maraia", "Laudmilla", "Alice-anne"]

#Verifica se exite um elemento igual a Nicolly na lista nomes e retorna um booleano
print(f"Existe o nome Nicolly na lista: {'Nicolly' in nomes}")
print()
```

```python
#métodos de listas


#Sort: ordena a lista em ordem crescente
nomes = ["Moisés", "Nicolly", "Maraia", "Laudmilla", "Alice-anne"]
bumer = [22, 23, 64, 5, 35, 2]

print(f"Nomes: {nomes}")
nomes.sort()
print(f"Nomes em ordem alfabética, usando sort(): {nomes}")
print()


print(f"Números: {bumer}")
bumer.sort()
print(f"Números em ordem crescente, usando sort(): {bumer}")
print()





#Reverse: retorna a lista ao contrário
bumer.reverse()
print(f"Lista invertida com reverse(): {bumer}")
print()

#Podemos usar o sort com o argumento reverse = True, para ordenamos em ordem decrescente:

nomes.sort(reverse = True)
print(nomes)
print()






#Count: conta o número de ocorrências de um elemento
nomes4 = ["Moisés", "Nicolly", "Maraia", "Laudmilla", "Alice-anne", "Moisés", "Laudmilla", "Laudmilla"]
print(f"Laudmilla aparece {nomes4.count('Laudmilla')} vezes!")
print()






#Index: retorna o índice da primeira ocorrência de um elemento
nomes5 = ["Moisés", "Nicolly", "Maraia", "Laudmilla", "Alice-anne", "Moisés", "Laudmilla", "Laudmilla"]
print(f"A primeira ocorrência de Laudmilla é no índice {nomes5.index('Laudmilla')}")
```
